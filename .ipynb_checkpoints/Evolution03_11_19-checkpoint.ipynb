```python
def allel(_size, maxdev):
    points=np.linspace(0,10,_size)/maxdev
    l=np.zeros(_size)
    for i in range(1,len(points)):
        dev=np.random.uniform(-1,1)/maxdev
        if l[i-1]>=1:
            l[i]=1
        elif l[i-1]<=-1:
            l[i]=-1
        else:
            l[i]=l[i-1]+dev
    return points, l
```

```python
f=allel(1000,1)
```

```python
plt.plot(f[0],f[1])
plt.show()
```
